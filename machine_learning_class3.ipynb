```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/NFLX.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
from statsmodels.tsa.arima.model import ARIMA

#fit model
model = ARIMA(df['High'], order=(5,1,0))
model_fit = model.fit()

#make prediction
forcast = model_fit.forecast(steps=12)
print(forcast)
```
